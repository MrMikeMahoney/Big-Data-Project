```python
##Creates list of features to pipe into estimator
features = x_train.columns
tf_variables =[]
for index in features:
    tf_variables.append(tf.feature_column.numeric_column(str(index)))
```

```python
tf_variables
```

```python
print(x_train.shape) 
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
##pandas
input_func=tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=100,num_epochs=None,shuffle=False)
```

```python
model = tf.estimator.DNNClassifier(feature_columns=tf_variables, hidden_units=[1024,512,256,128],optimizer=tf.train.ProximalAdagradOptimizer(
      learning_rate=0.01),dropout=.1,n_classes=303,activation_fn=tf.nn.relu)
```

```python
start_time = datetime.now()
model.train(input_fn = input_func, steps=30000)
end_time = datetime.now()
print("Duration: {}".format(end_time - start_time))
```

```python
#pandas
pred_fn = tf.estimator.inputs.pandas_input_fn(x=x_test,batch_size=len(x_test),shuffle=False)
```

```python
predictions = list(model.predict(input_fn=pred_fn))
```

```python
final_preds = []
for pred in predictions:
    final_preds.append(pred['class_ids'][0])
```

```python
print(classification_report(y_test,final_preds))
```
